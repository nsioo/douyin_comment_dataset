```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
result2 = pd.read_csv('../result_v2_process.csv', encoding='utf-8-sig')
result3 = pd.read_csv('../result_v3_process.csv', encoding='utf-8-sig')


def process(df):
    ff = pd.DataFrame(columns=['word', 'number'])
    s = {}
    for w in df['word']:
        if not w in s:
            s[w] = 1
        else:
            s[w] += 1
    
    words = list(s.keys())
    numbers = list(s.values())

    ff['word'] = words
    ff['number'] = numbers
    ff.sort_values(by=['number'], ascending=False, inplace=True)
    ff.reset_index(inplace=True, drop=True)
    return ff

uni_result_v2 = process(result2)
uni_result_V3 = process(result3)
```

```python

index2 = list(uni_result_v2.index)[:100]
index3 = list(uni_result_v2.index)[:100]
number2 = list(uni_result_v2.number)[:100]
number3 = list(uni_result_V3.number)[:100]


def graph_line(index2, index3, number2, number3):
    plt.figure(figsize=(8, 8))
    plt.plot(index2, number2, '-', color='red', label='result2',)
    plt.plot(index3, number3, '-', color='blue', label='result3')
    
    plt.legend(loc='best')
    # plt.title('The differences of domain specific lexicon with different seed words')
    # plt.xlabel('value of domain specific lexicon')
    plt.show()


graph_line(index2, index3, number2, number3)
```

```python

seed_word = pd.read_csv('../cn_seed_v2_15.csv', encoding='utf-8-sig')
seed = seed_word['word']
taget2 = uni_result_v2.loc[index2, 'word']
taget3 = uni_result_V3.loc[index3, 'word']


cmb2 = set(taget2) & set(seed)
cmb3 = set(taget3) & set(seed)

print('top100 from result2:', set(taget2))
print('top100 from result3:', set(taget3))
print(set(seed))
print(cmb2)
print(cmb3)

rate2 = len(cmb2) / len(seed)
rate3 = len(cmb3) / len(seed)

print('the rate of cmb with seed words')
print('--'*80)
print('the rate of result2 :{},  the rate of result3 : {}'.format(rate2, rate3))
```
